```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R=np.linspace(5,40,1000)
d_numeric = 0.959651*np.log(R-0.168581)+0.403843
d_numeric_max = (0.959651-0.1)*np.log(R-(0.168581+1.4))+(0.403843-0.38)
d_model = 2*np.log(R/2+1)-2*R/(R+2)
plt.plot(R,d_numeric,label="numeric best fit")
plt.plot(R,d_numeric_max,label="numeric fit 2")
plt.plot(R,d_model,label="model")
plt.legend()
plt.ylabel('d')
plt.xlabel('R')
plt.title('String Separation Model vs Numerics')
plt.savefig('String Separation Model vs Numerics.png')
```
