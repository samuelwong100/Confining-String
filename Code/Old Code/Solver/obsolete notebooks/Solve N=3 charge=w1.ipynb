Try to simulate SU(3) charge w1 (N-ality=1) on a small grid (10 by 10) and only 150 loops. From BPS solution, it is clear that the grid is too small for the BPS to properly relax. This will lead to a bad 2D simulation as well as not enough space is in the middle for it to relax to inertmediate vacuum. However, with only 150 loops, we can't see this, except notice that the error looks like it can go lower.

```python
sol = Solver_Full_Grid(N=3,charge_arg="w1",bound_arg="x1",L=10,w=10,h=0.1,R=5,tol=1e-30,max_loop=150,x0="BPS",diagnose=True)
```

With more loops (700 instead of 150), it is clear from the phi's that the space for the intermediate vacuum is very small. Consequently, the hole in the potential energy diagram is smaller.

```python
sol = Solver_Full_Grid(N=3,charge_arg="w1",bound_arg="x1",L=10,w=10,h=0.1,R=5,tol=1e-30,max_loop=700,x0="BPS",diagnose=True)
```

Introducing Continue_Solver technology, we simulate the 1000 loops using the 700 loops result, instead of starting over. In the comparison between middle slice and ideal BPS, it is clear that something isn't right. Although the location of peaks can shift (we just naively pasted together two BPS solutions), the height should be similar. But here, the slice height is quite different from the BPS height.

```python
Continue_Solver_Full_Grid("CS(N=3,charge=w1,bound=x1,L=10,w=10,h=0.1,R=5,tol=1e-30,max_loop=700,x0=BPS,method=full grid)",1000)
```

So we try to simulate it on a bigger grid (30 by 30) and two times farther apart (10). This time, it looks much better with more space in the middle for the vacuum to form. The log graph of the error looks like it can still go lower after 300 loops though.

```python
sol = Solver_Full_Grid(N=3,charge_arg="w1",bound_arg="x1",L=30,w=30,h=0.1,R=10,tol=1e-30,max_loop=300,x0="BPS",diagnose=True)
```

We use the continue_solver function to extend the simulation to 1000. This time the log graph finally converges (it looks like 400 was sufficient already). The double string potential energy looks really good. And as a great check, the middle slice and BPS almost looks identical except slightly shifted.

```python
Continue_Solver_Full_Grid("CS(N=3,charge=w1,bound=x1,L=30,w=30,h=0.1,R=10,tol=1e-30,max_loop=300,x0=BPS,method=full grid)",1000)
```
