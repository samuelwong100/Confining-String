```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def linear_model(x,m,b):
    return m*x + b
```

```python
R=np.array([10,15,20,25,30])
E=np.array([137.37,190.47,241.27,296.06,348.53])
potp, pcov = curve_fit(linear_model,xdata=R,ydata=E)
m,b = potp #slope and y-intercept
dm = np.sqrt(pcov[0][0]) #standard deviation of slope
```

```python
plt.figure()
plt.scatter(x=R,y=E)
x = np.linspace(1,35,1000)
plt.plot(x,linear_model(x,m,b))
plt.xlabel("R")
plt.ylabel("Energy")
plt.title("Energy vs Distance")
```

```python
print("E/R=",m,"+/-",dm)
```
