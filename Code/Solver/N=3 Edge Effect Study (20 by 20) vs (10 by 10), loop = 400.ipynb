```python
import numpy as np
import matplotlib.pyplot as plt
from Solution_Viewer import Solution_Viewer
from Solver_Full_Grid import Solver_Full_Grid, Continue_Solver_Full_Grid
```

```python
R_list_1 = []
energy_list_1 = []
R_list_2 = []
energy_list_2 = []

for R in range(2,10):
    sol = Continue_Solver_Full_Grid(
        "CS(N=3,charge=w1,bound=x1,L=10,w=10,h=0.1,R={},tol=1e-30,max_loop=150,x0=BPS,method=full grid)".format(R), max_loop=400)
    sol.plot_potential_energy_density()
    R_list_2.append(R)
    energy_list_2.append(sol.get_energy())
R_2 = np.array(R_list_2)
energy_2 = np.array(energy_list_2)

for R in range(2,10):
    sol = Continue_Solver_Full_Grid(
        "CS(N=3,charge=w1,bound=x1,L=20,w=20,h=0.1,R={},tol=1e-30,max_loop=150,x0=BPS,method=full grid)".format(R), max_loop=400)
    sol.plot_potential_energy_density()
    R_list_1.append(R)
    energy_list_1.append(sol.get_energy())
R_1 = np.array(R_list_1)
energy_1 = np.array(energy_list_1)
```

```python
plt.figure()
plt.scatter(R_1,energy_1,color="blue",s=5,label="grid=(20,20)")
plt.scatter(R_2,energy_2,color="red",s=5,label="grid=(10,10)")
plt.ylabel('Energy')
plt.xlabel("R")
plt.legend()
plt.title("N=3 Edge Effeect")
plt.savefig("N=3_Edge_Effect.png",dpi=300)
plt.show()
```

```python
plt.figure()
plt.scatter(R_1,energy_1,color="blue",s=5,label="grid=(20,20)")
plt.scatter(R_2,energy_2+45,color="red",s=5,label="grid=(10,10)")
plt.ylabel('Energy')
plt.xlabel("R")
plt.legend()
plt.title("N=3 Edge Effeect")
plt.savefig("N=3_Edge_Effect_Shifted.png",dpi=300)
plt.show()
```
